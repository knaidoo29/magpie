# Tutorial 3: Monte Carlo Weighted Remapping - Cube2Shell

```python
import numpy as np
import matplotlib.pylab as plt
import healpy as hp
import magpie
```

```python
# Initialise magpie Cub2Shell class
c2s = magpie.mc.Cube2Shell()

# Setup the cube grid
c2s.setup_cube(0., 2., 10, 0., 2., 10, 0., 2., 10)

# We are going to just assign each point an ascending integer value.
f = np.zeros(np.shape(c2s.x3d))
count = 0.
for i in range(0, len(c2s.x3d)):
    for j in range(0, len(c2s.x3d[0])):
        for k in range(0, len(c2s.x3d[0][0])):
            f[i, j, k] = count
            count += 1.
```

```python
# Construct spherical polar coordinate shells
nside = 16
c2s.setup_polar_lin(0., 2., 10, nside, center=[0., 0., 0.])
# Calculate monte carlo weights for remapping
c2s.get_weights()
# The weights for a particular remapping can be stored.
```

```python
# Now, let's remap the function from a linear cartesian grid to spherical polar coordinates
f_sphere = c2s.remap(f)
```

```python
for i in range(0, len(f_sphere)):
    hp.mollview(f_sphere[i])
```

```python
# We can also change the center, moving it to (x, y) = (1, 1, 0)
nside = 16
c2s.setup_polar_lin(0., 2., 10, nside, center=[1., 1., 0.])
# Calculate monte carlo weights for remapping
c2s.get_weights()
# The weights for a particular remapping can be stored.
```

```python
# Now, let's remap the function from a linear cartesian grid to spherical polar coordinates
f_sphere = c2s.remap(f)
```

```python
for i in range(0, len(f_sphere)):
    hp.mollview(f_sphere[i])
```
